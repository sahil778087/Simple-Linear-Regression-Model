# 📈 Salary Prediction: Simple Linear Regression Analysis

### 🎯 Project Overview
This project implements a **Simple Linear Regression** model to analyze and predict the relationship between **Years of Experience** and **Salary**. By training on a dataset of professional earnings, the model learns to estimate the financial trajectory of a career based on tenure.

---

### 🧠 Mathematical Foundation
The model assumes a linear relationship represented by the following equation:

$$y = \beta_0 + \beta_1 X + \epsilon$$

Where:
* **$y$**: Dependent Variable (**Salary**)
* **$X$**: Independent Variable (**Years of Experience**)
* **$\beta_0$**: The **Intercept** (The predicted salary for 0 years of experience)
* **$\beta_1$**: The **Slope/Coefficient** (The average salary increase per year of experience)
* **$\epsilon$**: The **Error Term** (Residuals)

The target variable, **Salary**, is a continuous numerical value. Therefore, we will perform a **Regression** analysis rather than Classification. 

Since we are modeling the relationship between a single independent variable ($X$) and a dependent variable ($y$), we define the linear relationship as:

$$y = \beta_0 + \beta_1 X + \epsilon$$

Where:
* $y$ is the **Salary** (Target)
* $X$ is the **Years of Experience** (Feature)
* $\beta_0$ is the intercept
* $\beta_1$ is the slope (coefficient)
